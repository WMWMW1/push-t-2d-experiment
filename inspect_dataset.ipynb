```python
from datasets import load_dataset

# 加载 train split
ds = load_dataset("lerobot/pusht_image", split="train")

# 比如看第 0 帧（episode 1）的 action
action0 = ds["action"][0]
print("action[0] =", action0)  # 例如 [236.273, 313.15808]

# 如果你想检查第 N 帧：
N = 100
actionN = ds["action"][N]
print(f"action[{N}] =", actionN)

```

```python
# 假设每集正好 100 帧
EP_LEN = 100
episode = 1  # 第 2 集
start = episode * EP_LEN
end   = start + EP_LEN

for idx in range(start, end):
    print(f"frame {idx}: action = {ds['action'][idx]}")


```

```python
eps_idx = ds["episode_index"]   # list of ints

# 3) 统计不同的 index 数量
n_eps = len(set(eps_idx))
print(f"Total episodes in train split: {n_eps}")
```

```python
from collections import Counter
counts = Counter(eps_idx)
for ep, length in counts.items():
    print(f"Episode {ep:02d} length = {length} transitions")
```
